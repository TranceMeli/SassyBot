# Chatbot guide




#### 1. importing libraries

```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
```

#### 2. creating the chatbot
name of the bot<br>
store the data in a database<br>
save trained data<br>

```python
chatbot = ChatBot(
    'Charlie',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite///database.sqlite3'
)
```

tells the bot which "logic brains" it should use to decide on responses<br>
ChatterBot has logic adapters, each one is like a module that handles a certain type of question<br>
you can list one or many adapters, and the bot will try them in order until it finds a good response

```python
bot = ChatBot(
    'Charlie',
    logic_dapters=[
        'chatterbot.logic.BestMatch',
        'chatter ot.logic.TimeLogicAdapter'],
)
```

### 3. creating the trainer
trainer is now an object you can use to teach the bot a new conversations<br>
ListTrainer expects lists of statements or pair of statements

```python
trainer = ListTrainer(chatbot)
```

#### 4. training the bot
this trains the bot with a sequence of statements

```python
trainer.train(
    [
        "Hello",
        "Hi there!",
        "How are you doing?",
        "I'm doing great.",
        "That is good to hear",
        "Thank you",
        "You're welcome",
        "You are amazing",
        "You are wonderful"
    ]
)
```

#### 5. testing the initial response
bot searches database for matching response
